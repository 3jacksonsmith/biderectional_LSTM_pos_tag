### Functions

```python
def read_data(path_data):

    sentences = []

    with open (path_data) as myfile:
        for line in myfile:
            lineFile = line.split(".")
            line = str.split(str.strip(line), ' ')
            item = []

            for word in line:
                string_word = str.split(str.strip(word), '_')
                item.append(string_word)

            sentences.append(item)
    return sentences
```

```python
import numpy as np
def to_categorical(sequences, categories):
    cat_sequences = []
    for s in sequences:
        cats = []
        for item in s:
            cats.append(np.zeros(categories))
            cats[-1][item] = 1.0
        cat_sequences.append(cats)
    return np.array(cat_sequences)
```

```python
def separe_sentence_tags(tagged_sentences): 
    import numpy as np
 
    sentences, sentence_tags =[], [] 

    for tagged_sentence in tagged_sentences:
        sentence, tags = zip(*tagged_sentence)
        sentences.append(np.array(sentence))
        sentence_tags.append(np.array(tags))
    return sentences, sentence_tags
```

```python
def create_word2index_tag2index(all_sentences, all_tags):
    words, tags = set([]), set([])

    for s in all_sentences:
        for w in s:
            words.add(w.lower())

    for ts in all_tags:
        for t in ts:
            tags.add(t)

    word2index = {w: i + 2 for i, w in enumerate(list(words))}
    word2index['-PAD-'] = 0  # The special value used for padding
    word2index['-OOV-'] = 1  # The special value used for OOVs

    tag2index = {t: i + 1 for i, t in enumerate(list(tags))}
    tag2index['-PAD-'] = 0  # The special value used to padding
    return word2index, tag2index
```

```python
def create_sentences_tags_to_number(sentences_nlp, tags_nlp, word2index, tag2index):
    sentences_X, tags_y = [], []
 
    for s in sentences_nlp:
        s_int = []
        for w in s:
            try:
                s_int.append(word2index[w.lower()])
            except KeyError:
                s_int.append(word2index['-OOV-'])

        sentences_X.append(s_int)

    for s in tags_nlp:
        tags_y.append([tag2index[t] for t in s])
        
    
    print(sentences_X[0])
    print(tags_y[0])
    
    return sentences_X, tags_y
```

### Create Train

```python
sentences_train = read_data("./macmorpho-v3/macmorpho-train.txt")
```

```python
sentences_nlp_train, tags_nlp_train = separe_sentence_tags(sentences_train)
```

```python
def create_word2index_tag2index(all_sentences, all_tags):
    words, tags = set([]), set([])

    for s in all_sentences:
        for w in s:
            words.add(w.lower())

    for ts in all_tags:
        for t in ts:
            tags.add(t)

    word2index = {w: i + 2 for i, w in enumerate(list(words))}
    word2index['-PAD-'] = 0  # The special value used for padding
    word2index['-OOV-'] = 1  # The special value used for OOVs

    tag2index = {t: i + 1 for i, t in enumerate(list(tags))}
    tag2index['-PAD-'] = 0  # The special value used to padding
    return word2index, tag2index
```

```python
word2index_train, tag2index_train = create_word2index_tag2index(sentences_nlp_train, tags_nlp_train)
```

### Create Model and Validate

```python
from keras.preprocessing.sequence import pad_sequences
 
train_sentences_X = pad_sequences(train_sentences_X, maxlen=MAX_LENGTH_train, padding='post')
test_sentences_X = pad_sequences(test_sentences_X, maxlen=MAX_LENGTH_train, padding='post')
train_tags_y = pad_sequences(train_tags_y, maxlen=MAX_LENGTH_train, padding='post')
test_tags_y = pad_sequences(test_tags_y, maxlen=MAX_LENGTH_train, padding='post')
 
print(train_sentences_X[0])
print(test_sentences_X[0])
print(train_tags_y[0])
print(test_tags_y[0])
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, InputLayer, Bidirectional, TimeDistributed, Embedding, Activation
from keras.optimizers import Adam
 
 
model = Sequential()

model.add(InputLayer(input_shape=( MAX_LENGTH_train, )))
model.add(Embedding(len(word2index_train), 128))
model.add(Bidirectional(LSTM(256, return_sequences=True)))
model.add(TimeDistributed(Dense(len(tag2index_train))))
model.add(Activation('softmax'))
 
model.compile(loss='categorical_crossentropy',
              optimizer=Adam(0.001),
              metrics=['accuracy'])
 
model.summary()
```

```python
model.fit(train_sentences_X, to_categorical(train_tags_y, len(tag2index_train)), batch_size=128, epochs=8, validation_split=0.2)
```

```python
scores = model.evaluate(test_sentences_X, to_categorical(test_tags_y, len(tag2index_train)))
print(f"{model.metrics_names[1]}: {scores[1] * 100}")   # acc: 99.09751977804825
```

```python
def logits_to_tokens(sequences, index):
    token_sequences = []
    for categorical_sequence in sequences:
        token_sequence = []
        for categorical in categorical_sequence:
            token_sequence.append(index[np.argmax(categorical)])
 
        token_sequences.append(token_sequence)
 
    return token_sequences
```

### improvement

```python
from keras import backend as K
 
def ignore_class_accuracy(to_ignore=0):
    def ignore_accuracy(y_true, y_pred):
        y_true_class = K.argmax(y_true, axis=-1)
        y_pred_class = K.argmax(y_pred, axis=-1)
 
        ignore_mask = K.cast(K.not_equal(y_pred_class, to_ignore), 'int32')
        matches = K.cast(K.equal(y_true_class, y_pred_class), 'int32') * ignore_mask
        accuracy = K.sum(matches) / K.maximum(K.sum(ignore_mask), 1)
        return accuracy
    return ignore_accuracy
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, InputLayer, Bidirectional, TimeDistributed, Embedding, Activation
from keras.optimizers import Adam
 
 
model = Sequential()
model.add(InputLayer(input_shape=(MAX_LENGTH_train, )))
model.add(Embedding(len(word2index_train), 128))
model.add(Bidirectional(LSTM(256, return_sequences=True)))
model.add(TimeDistributed(Dense(len(tag2index_train))))
model.add(Activation('softmax'))
 
model.compile(loss='categorical_crossentropy',
              optimizer=Adam(0.001),
              metrics=['accuracy', ignore_class_accuracy(0)])
 
model.summary()
```

```python
model.fit(train_sentences_X, to_categorical(train_tags_y, len(tag2index_train)), batch_size=128, epochs=8, validation_split=0.2)
```

```python
scores = model.evaluate(test_sentences_X, to_categorical(test_tags_y, len(tag2index_train)))
print(f"{model.metrics_names[1]}: {scores[1] * 100}")   # acc: 99.09751977804825
```
